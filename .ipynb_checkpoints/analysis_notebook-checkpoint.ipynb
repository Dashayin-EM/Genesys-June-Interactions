# Genesys June Interactions Analysis

This notebook loads the exported CSV, runs the textual analysis from `analyzer.py`, and visualises the main metrics.

**Prerequisites**
- Python 3.11+
- Packages: `pandas`, `matplotlib`, `seaborn` (install with `pip install matplotlib seaborn`).
- The notebook lives in the agent's artifact folder, but it explicitly adds the project root to `sys.path` so it can import `loader` and `analyzer`.

```python
import os, sys

# Absolute path to the workspace root (where loader.py / analyzer.py live)
project_root = r"C:/Users/dashayin/Desktop/Work/Genesys June Interactions"
if project_root not in sys.path:
    sys.path.insert(0, project_root)
print('Project root added to sys.path:', project_root)

# Verify imports
try:
    import loader
    import analyzer
    import config as cfg
    import matplotlib.pyplot as plt
    import seaborn as sns
    %matplotlib inline
    sns.set(style='whitegrid', palette='deep')
    print('All modules imported successfully')
except Exception as e:
    print('Import error:', e)
    raise
```

```python
# Load the CSV using the project's loader
csv_path = os.path.join(project_root, 'data', 'interactions.csv')
df = loader.load_data(csv_path)
df.head()
```

```python
# Run the textual analysis (prints summary)
analyzer.run_analysis(df)
```

```python
# 1️⃣ Daily interaction count for June 2026
if 'Parsed_Timestamp' in df.columns:
    june_df = df[df['Parsed_Timestamp'].dt.month == 6]
    daily = june_df.groupby(june_df['Parsed_Timestamp'].dt.date).size()
    plt.figure(figsize=(10,4))
    daily.plot(kind='bar', color='#1F4E78')
    plt.title('Daily Interaction Count – June 2026')
    plt.xlabel('Date')
    plt.ylabel('Number of Interactions')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 2️⃣ Top 15 system error codes (Updated to use config mapping)
if cfg.ERROR_CODE_COL in df.columns:
    # Filter out empty strings first
    err_df = df[df[cfg.ERROR_CODE_COL] != '']
    top_errors = err_df[cfg.ERROR_CODE_COL].value_counts().nlargest(15)
    plt.figure(figsize=(10,4))
    top_errors.plot(kind='bar', color='#d9534f')
    plt.title('Top 15 System Error Codes')
    plt.xlabel('Error Code')
    plt.ylabel('Occurrences')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# 3️⃣ Interaction duration distribution (seconds)
if 'Duration_Seconds' in df.columns:
    plt.figure(figsize=(8,4))
    # Filter for > 0 to remove zero-second abandoned blips skewing the distribution
    sns.histplot(df[df['Duration_Seconds'] > 0]['Duration_Seconds'].dropna(), bins=30, kde=True, color='#2a69a4')
    plt.title('Interaction Duration Distribution')
    plt.xlabel('Duration (seconds)')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
# 4️⃣ Top 10 queues by interaction count (Updated to use exploded config mapping)
if cfg.QUEUE_COL in df.columns:
    exp_queue = analyzer.explode_multivalued_column(df, cfg.QUEUE_COL)
    if not exp_queue.empty:
        top_queues = exp_queue['split_val'].value_counts().nlargest(10)
        plt.figure(figsize=(10,4))
        top_queues.plot(kind='bar', color='#1F4E78')
        plt.title('Top 10 Queues – Interaction Count')
        plt.xlabel('Queue')
        plt.ylabel('Interactions')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

```python
# 5️⃣ Top 10 agents by interaction count (Updated to use exploded config mapping)
if cfg.AGENT_COL in df.columns:
    exp_agent = analyzer.explode_multivalued_column(df, cfg.AGENT_COL)
    if not exp_agent.empty:
        top_agents = exp_agent['split_val'].value_counts().nlargest(10)
        plt.figure(figsize=(10,4))
        top_agents.plot(kind='bar', color='#1F4E78')
        plt.title('Top 10 Agents – Interaction Count')
        plt.xlabel('Agent')
        plt.ylabel('Interactions')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```
